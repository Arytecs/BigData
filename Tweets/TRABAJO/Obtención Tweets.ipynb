Descarga de tweets

```python
import tweepy
import json

from twitter_secret import *

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

API = tweepy.API(auth)

print (API)
```

```python
filename_train = "../TRABAJO/TASS2014_training_polarity.txt"

with open(filename_train) as fh:
    id_list_train = [x.split() for x in fh.read().split('\n')[:-1]]

print (len(id_list_train))
print ('-'*20)
all_tweets_train = []
c = 0
while c < len(id_list_train):
    ids = [x for x,y in id_list_train[c:c+100]]
    tweets = API.statuses_lookup(ids)
    all_tweets_train += tweets
    #print (len(ids), len(tweets))
    c += 100

print ('-'*20)
print (len(all_tweets_train))
```

```python
import re
er = re.compile('\s')

def save_tweets(filename, tweets):
    with open(filename, 'w', encoding="utf8") as fh:
        for tweet in tweets:
            tid = str(tweet.id)
            text = er.sub(' ', tweet.text) # eliminamos separadores "raros" (opcion 2)
            line = '\t'.join((tid, text))
            fh.write(line + '\n')
            
def load_tweets(filename):
    tweets = []
    for line in open(filename):
        #print ("'"+line+"'")
        #print (line.split('\t'))
        tid, user, text = line.split('\t')
        tweets.append({'id':tid, 'user':user, 'text':text})
    return tweets
```

```python
save_tweets('tweets_train.txt',all_tweets_train)
```

```python
filename = "../TRABAJO/TASS2014_test_ids.txt"

with open(filename) as fh:
    id_list_test = fh.readlines()

print (len(id_list_test))
print ('-'*20)
all_tweets_test = []
c = 0
while c < len(id_list_test):
    ids = [x.strip() for x in id_list_test[c:c+100]]
    tweets = API.statuses_lookup(ids)
    all_tweets_test += tweets
    #print (len(ids), len(tweets))
    c += 100

print ('-'*20)
print (len(all_tweets_test))
```

Hacemos merge del conjunto de train para guardar los resultados

```python
import pandas as pd
tweets_text = pd.read_csv('tweets_train.txt', sep="\t", header=None, encoding="utf8")
tweets_result = pd.read_csv('TASS2014_training_polarity.txt', sep="\t", header=None)

tweets_text = pd.DataFrame(all_tweets_train)
df = pd.DataFrame()
df["id"] = [i.id for i in tweets_text[0]]
df['text'] = [i.text for i in tweets_text[0]]
df.columns =["id","text"]
tweets_result.columns =["id","result"]
inner_data=pd.merge(df, tweets_result, on='id', how='inner')

print(inner_data.shape)
```

Y guardamos los datos unidos con los resultados en un nuevo fichero

```python
filename = "tweets_train_result.txt"
with open(filename, 'w', encoding="utf8") as fh:
    for i in range(0,len(inner_data)):
        #print(i)
        result = inner_data.loc[i][2]
        idTweet=inner_data.loc[i][0]
        text=inner_data.loc[i][1].replace("\n"," ").replace("\r", " ")
        line = '\t'.join((str(idTweet), text, result))
        fh.write(line + '\n')
        
    
```
